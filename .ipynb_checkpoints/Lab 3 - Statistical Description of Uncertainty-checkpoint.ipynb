## P3.1 
Using Python, verify the following calculation using Python. 
<br><br>
$$ \bar{x} = \frac{2.02 + 3.01 + 2.63 + 3.12 + 2.65 + 2.13 + 1.66 + 2.52 + 1.86 + 2.31}{10} = 2.39 $$
<br>

```python
# (a) ... by explicitly performing the sum from above
values = [2.02, 3.01, 2.63, 3.12, 2.65, 2.13, 1.66, 2.52, 1.86, 2.31]
mean_manual = sum(values) / 10.0
print (mean_manual)
```

```python
# (b) ...by Using the function mean inside of numpy
import numpy as np
mean_numpy = np.mean(values)
print (mean_numpy)
```

```python
# Verify that the two methods yield the same answer:
print (mean_numpy == mean_manual)
```

## P3.2
Using Python, calculate the standard deviation of the data from the previous section

##### Standard Deviation Formula
$$ \sigma = \sqrt{\frac{1}{N} \sum\limits_{i=1}^N (x - \bar{x})^2} $$

```python
# (a) explicitly calculating the sum in equation (3.6)
mean = np.mean(values)
std_manual = np.sqrt(sum([(x - mean)**2 for x in values]) / float(len(values)))
print (std_manual)
```

```python
# (b) using the function `std` inside of numpy.
std = np.std(values)
print (std)
```

The above results approximate the expected `0.46` for the standard deviation.

##### Take a second to jot down a paragraph about the difference between random and systematic error.

Random errors are errors that come from the nature of measuring. For example, when measuring the weight of various elements, there will always be error. Some error is negative (measuring an element to be to light) and some error is positive (measuring an element to be too heavy). Conjecturally, these will be evenly distributed (about the same number of negative errors as positive). Systematic error involves a consistent error unnaturally found while measuring. For example, the scale being used to measure may be off by 2g from reality. Clearly, systematic errors cannot be caluculated, only observed and compensated for.
